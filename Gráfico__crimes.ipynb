```python
import matplotlib.pyplot as plt

datas = [6, 7, 8, 9]
Alvorada = [14, 11, 16, 23]
Porto = [127, 100, 123, 133]
Viamão = [19, 18, 10, 12]

plt.figure(figsize=(10, 8)) 
plt.plot(datas, Alvorada, label='Alvorada', marker='o')
plt.plot(datas, Porto, label='Porto Alegre', marker='s')
plt.plot(datas, Viamão, label='Viamão', marker='d')

plt.title('Roubo de veiculos em 2023')
plt.xlabel('Meses')
plt.ylabel('Quantidade')
plt.grid(True)
plt.legend()

plt.show()
```

```python
import pandas as pd
import folium

dados = {
    'nome': ['Alvorada', 'Porto Alegre', 'Viamão'],
    'cidade': ['Alvorada', 'Porto Alegre', 'Viamão']
}

df = pd.DataFrame(dados)

coordenadas = {
    'Alvorada': [-29.999939, -51.070272],
    'Porto Alegre': [-30.0346, -51.2177],
    'Viamão': [-30.102932, -51.025887]
    
}

m = folium.Map(location=[-30.028033, -51.086126], zoom_start=12)


for index, row in df.iterrows():
    folium.Marker(location=coordenadas[row['cidade']], 
                  popup=row['nome']).add_to(m)

m
```
