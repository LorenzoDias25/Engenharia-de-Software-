```python
import matplotlib.pyplot as plt

# Datas de 2015 a 2020
datas = ['2015-01-01', '2016-01-01', '2017-01-01', '2018-01-01', '2019-01-01', '2020-01-01']

# Preços fictícios da Netflix para cada ano
precos_netflix = [100, 120, 150, 180, 200, 220]

# Criar um gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(datas, precos_netflix, marker='o', linestyle='-', color='b')
plt.title('Preço da Netflix (2015-2020)')
plt.xlabel('Ano')
plt.ylabel('Preço em USD')
plt.xticks(rotation=45)
plt.grid(True)

# Exibir o gráfico
plt.tight_layout()
plt.show()
```
